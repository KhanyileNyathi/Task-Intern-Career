```python
plt.plot(df.Goals,df.Shots,'o')
plt.title('Goals and Shots')
plt.show()
```

```python
#group for easy seperation by year
df['sep']=pd.cut(df['Year'],bins=5,labels=('2016','2017','2018','2019','2020'))
df
```

```python
#find the amount of goals scored each year
yearly_goals=df.groupby(['sep'])['Goals'].sum()
yearly_goals
```

```python
plt.plot(df.sep ,df.Goals)
plt.plot(df.sep,df.Shots)
plt.plot(df.sep ,df.Mins)
plt.title('Stats for every year')
plt.legend(['Goals','Shots Taken','Minutes Played'])
plt.xlabel('Year')
plt.ylabel('Goals Scored per year')
plt.show()
```

```python
#relative growth in goals scored
df.Goals/df.Goals.iloc[0]*100
```

```python
#compare the %
plt.plot(df.sep ,df.Goals/df.Goals.iloc[0]*100)
plt.plot(df.sep,df.Shots/df.Shots.iloc[0]*100)
plt.title('Percentage Goals And Shots Taken For Every Year')
plt.legend(['Goals','Shots Taken'])
plt.show()
```
